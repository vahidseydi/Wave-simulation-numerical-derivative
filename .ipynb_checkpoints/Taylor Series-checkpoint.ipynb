estimate $ cos(x) $ with taylor series

```python
import numpy as np
from math import *
import matplotlib.pyplot as plt
from matplotlib import animation,rc
from IPython.display import HTML
```

```python
xmax = 5                   # physical domain (m)
xmin = -5                    # physical domain (m)

nx = 100                        # number of samples    
dx = (xmax-xmin)/(nx-1)                # grid increment dx (m)
x  = np.linspace(xmin,xmax,nx)     # space coordinates
dx
```

```python
# Initialization of sin function
l = 2*pi         # wavelength
k = 2*pi/l    # wavenumber
```

```python
f = np.cos(k*x)
```

```python
# Plot sin function
plt.figure(figsize=(10,6))
plt.plot(x, f)
plt.title('Cos function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim((xmin, xmax))
plt.grid()
plt.show()
```

\begin{equation}
f(x+dx)= f(x) + f^{'}(x)\times dx +\frac{1}{2!} f^{''}(x)\times dx^{2}+...
\end{equation}

\begin{equation}
f(x) = f(x+dx) - ( f^{'}(x)\times dx +\frac{1}{2!} f^{''}(x)\times dx^{2}+...)
\end{equation}
<span style="color: red;">
just to remember the relation between two forms of the Taylor series! \
$ a = x+dx $ \
$=> dx = a-x = -(x-a)$\
$=> f(x) = f(a) + f^{'}(x)\times (x-a) +\frac{1}{2!} f^{''}(x)\times (x-a)^{2}+...$
</span>.

```python
Order = 20
ef =np.zeros([Order,len(x)])
for T in range(Order):
    for i,dx in enumerate(x):
        fx_plus_dx=0
        sign =1
        for t in range(0,T,2):
            fx_plus_dx += sign* np.power(dx,t) / factorial(t) 
            sign *= -1
        ef[T,i]=fx_plus_dx
```

```python
fig = plt.figure(figsize=(10,6))
ax = plt.axes(xlim=(xmin, xmax), ylim=(-2, 2)) 
ax.plot(x, f, lw=3)
line, = ax.plot([], [], '--r',lw=3)

def init():
    line.set_data([], [])
    return line,
def animate(i):    
    line.set_data(x,ef[i,:])
    line.set_label('order %d'%(i))
    return line,

anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=ef.shape[0], interval=200)

```
